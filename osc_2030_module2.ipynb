```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://github.com/matthewpecsok/osc_2030/raw/refs/heads/main/Colleges%20and%20Universities%20-%20OIS%202030.xlsx'

colleges_and_univ = pd.read_excel(url)

colleges_and_univ.head()
```

```python
math_std_by_state = colleges_and_univ.groupby('State')['Math SAT'].std(ddof=0) # 0 = population standard devation.

math_std_by_state
```

```python
plt.figure(figsize=(16, 6))
plt.bar(math_std_by_state.index, math_std_by_state.values)
plt.xlabel("State")
plt.ylabel("Standard Deviation of Math SAT Scores")
plt.title("Standard Deviation of Math SAT Scores by State")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
math_std_by_state_sorted = math_std_by_state.sort_values()

plt.figure(figsize=(10, 6))
plt.bar(math_std_by_state_sorted.index, math_std_by_state_sorted.values)
plt.xlabel("State")
plt.ylabel("Standard Deviation of Math SAT Scores")
plt.title("Standard Deviation of Math SAT Scores by State (Sorted)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
